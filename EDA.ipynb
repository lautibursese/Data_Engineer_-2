```python
var = "GrLivArea"
```

```python
data= pd.concat([df_train["SalePrice"],df_train[var]], axis=1)
```

```python
data.plot.scatter(x = var, y = "SalePrice", ylim = (0,800000))
```

```python
var2 = "TotalBsmtSF"
data2 = pd.concat([df_train["SalePrice"],df_train[var2]], axis=1)
data2.plot.scatter(x=var2, y="SalePrice", ylim = (0,800000))
```

```python
OverallQual = "OverallQual"
data3 = pd.concat([df_train["SalePrice"],df_train[OverallQual]], axis=1)
data3.plot.scatter(x=OverallQual, y = "SalePrice", ylim=(0,800000))
```

```python
f,ax = plt.subplots(figsize =(10,8))    #specific dimension to the graphics
fig = sns.boxplot(x = OverallQual,y="SalePrice", data=data3)
fig.axis(ymin = 0, ymax=800000)
```

```python
YearBuilt = "YearBuilt"
data4 = pd.concat([df_train["SalePrice"],df_train[YearBuilt]], axis=1)

f,ax = plt.subplots(figsize =(20,10)) 
fig = sns.boxplot(x = YearBuilt,y="SalePrice", data=data4)
fig.axis(ymin = 0, ymax=800000)
plt.xticks(rotation = 90)
plt.show()
```

```python
CorrmatMatrix = df_train.corr()
f,ax = plt.subplots(figsize =(12,9))
sns.heatmap(CorrmatMatrix, vmax = 1.1, square=True)
```

```python
k=10    
```

```python
cols = CorrmatMatrix.nlargest(k,"SalePrice")["SalePrice"].index # Gets the more related "K" values
cols
```

```python
cm = np.corrcoef(df_train[cols].values.T)   # Transposition
cm
```

```python
sns.set(font_scale=1.25)    # Change the format of the letters
```

```python
f,ax = plt.subplots(figsize =(12,9))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
```

```python
sns.set()
```
